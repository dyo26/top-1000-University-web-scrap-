```python
#import the python request library to query a website
import requests
#specify the url we want to scrape from
Link = "https://en.wikipedia.org/wiki/List_of_top_1000_universities_in_the_world"
#convert the web page to text
Link_text = requests.get(Link).text
print(Link_text)
```

```python
#import BautifulSoup library to pull data out of HTML and XML files
from bs4 import BeautifulSoup
#to convert Link_text into a BeautifulSoup Object
soup = BeautifulSoup(Link_text, 'lxml')
print(soup)
```

```python
#make the indentation proper
print(soup.prettify())
```

```python
#To take a look at the title of the web page
print(soup.title)
```

```python
#Only the string not the tags
print(soup.title.string)
```

```python
#First <a></a> tag
soup.a
```

```python
#all the <a> </a> tags
soup.find_all('a')
```

```python
#Fetch all the table tags
rows = soup.find_all('tr')
print(rows)
```

```python
for row in rows:
    each_row= row.findAll('th')
    print(each_row)
```

```python
lists_of_rows = []
for row in rows:
    each_row= row.findAll('td')
    str_row= str(each_row)
    row_text = BeautifulSoup(str_row, "lxml").get_text()
    lists_of_rows.append(row_text)
print(lists_of_rows)
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame(lists_of_rows[2:])
print(data)
```

```python
data.to_csv('Univ.csv', encoding='utf-8', index=False)
```

```python
df = pd.read_csv('Univ.csv')
df
```

```python
df.info()
```

```python
new = df["0"].str.split(",", n=1, expand = True)
new
```

```python
new[1]
```

```python
country = new[1].str.split(",", n=1, expand = True)
country
```

```python
df['University'] = new[0]
df['Location'] = country[0]
df
```

```python
df.drop(['0'], axis=1, inplace= True)
df
```

```python
import re
```

```python
def cleaning_time (x):
    y = x.rstrip("\n")
    z = re.sub(r'[[]', "", y)
    return z
```

```python
df1 = df.copy()
```

```python
df1['University'] = df1['University'].apply(lambda x: cleaning_time(x))
df1
```

```python
df1['Location'] = df1['Location'].apply(lambda x: cleaning_time(x))
df1
```

```python
df1.to_csv('Univ2.csv', encoding='utf-8', index=False)
```
